```python
def brightest(picture): 
    """add up each pixel;s color values to approximate its overall luminance"""
    xsize, ysize = picture.size
    bx, by, max_val = 0, 0, 0
    for x in range(xsize): 
        for y in range(ysize): 
            r,g,b = pic.getpixel((x,y))
            if r+b+g > max_val: 
                bx, by, max_val = x, y, r+g+b
    return (bx, by), max_val
```

```python
from time import time
def elapsed(func, picture):
    """Takes a function and a picture as arguments, applies the function to the picture, and returns the elapsed time along with whatever the function itself returned."""
    start = time()
    result = func(picture)
    return time() - start, result
```

```python
print elapsed(brightest,pic)
```

```python
def faster(picture): 
    """This function, 'faster', uses 'picture.getdata' to unpack the row-and-column representation of the image to create a vector of pixels, and then it loops over that."""
    max_val = 0
    for (r,g,b) in picture.getdata():
        if r+g+b > max_val: 
            max_val = r + g + b
    return max_val
```

```python
print elapsed(faster,pic)
```

```python
def inbetween(picture):
    xsize, ysize = picture.size
    temp = picture.load()
    bx, by, max_val = 0, 0, 0
    for x in range(xsize): 
        r,g,b = temp[x,y]
        if r+g+b > max_val: 
            bx, by, total = x,y,r+g+b
    return (bx, by), total
```

```python
print elapsed(inbetween,pic)
```

## Finding Stars

```python
def monochrome(picture, threshold):
    """loops over the pixels in the loaded image, 
    replacing the RGB values of each with either 
    black or white depending on whether its total 
    luminance is above or below some threshold 
    passed in by the user"""
    black = (0, 0, 0)
    white = (255, 255, 255)
    xsize, ysize = picture.size
    temp = picture.load()
    for x in range(xsize):
        for y in range(ysize):
            r,g,b = temp[x,y]
            if r+g+b >= threshold: 
                temp[x,y] = black
            else:
                temp[x,y] = white
```

```python
#get another copy to convert to B/w
bwpic = Image.open(StringIO(response.content))

#Remember, this threshold is a scalar, not a RGB triple
#We're looking for pixels whose total color value is 600 or greater
monochrome(bwpic, 200+200+200)
plt.imshow(bwpic);
```

```python
BLACK = (0, 0, 0)
RED = (255, 0, 0)
def count(picture): 
    """Scan the image top to bottom and left to right using a nested loop. When black pixel is found, incriment the count, then call the fill function to fill in all the pixels conected to that one."""
    xsize, ysize = picture.size
    temp = picture.load()
    result = 0
    for x in range(xsize):
        for y in range(ysize):
            if temp[x,y] == BLACK: 
                result += 1
                fill(temp,xsize,ysize,x,y)
    return result
```

```python
def fill(picture, xsize, ysize, xstart, ystart):
    """keep a list of pixels that need to be looked at, but haven't yet been filled in - a list of the (x,y) coordinates of pixels that are neighbors of ones we have already examined.  Keep looping until there's nothing left in this list"""
    queue = [(xstart,ystart)]
    while queue:
        x,y,queue = queue[0][0], queue[0][1], queue[1:]
        if picture[x,y] == BLACK:
            picture[x,y] = RED
            if x > 0:
                queue.append((x-1,y))
            if x < (xsize-1):
                queue.append((x+1,y))
            if y > 0:
                queue.append((x, y-1))
            if y < (ysize-1):
                queue.append((x, y+1))
                
```

```python
count(bwpic)
```

```python
def fillrec(picture, xsize, ysize, x, y): 
    """Each call to 'fillrec' takes care of one pixel, then calls 'fillrec' again to take care of the neighbors"""
    if picture[x,y] != BLACK:
        return
    picture[x,y] = RED
    if x > 0: 
        fillrec(picture, xsize, ysize, x-1, y)
    if x < (xsize-1): 
        fillrec(picture, xsize, ysize, x, x+1, y)
    if y > 0: 
        fillrec(picture, xsize, ysize, x, y-1)
    if y < (ysize-1): 
        fillrec(picture, xsize, ysize, x, y+1)
```

```python
print elapsed(count,bwpic)
```
