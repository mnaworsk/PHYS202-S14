```python
g = 9.8 #m/s^2
ell = 0.5 #m
def deriv(theta_current,t):  # return derivatives of the array theta
    theta = theta_current[0]
    omega = theta_current[1]
    dtheta = omega
    domega = -(g/ell)*np.sin(theta)
    return np.array([ dtheta, domega ])

from scipy.integrate import odeint
t = np.arange(0.,14.+0.01,0.01)

th0 = 0
th1 = np.pi / 4
th2 = np.pi / 2
th3 = 3*np.pi / 4
th4 = np.pi
omega0 = 0. #starting from rest, omega = 0
theta_init = np.array([th0, omega0]) # initial values, th0 from before = 20 degrees, expressed in radians
theta_init1 = np.array([th1, omega0])
theta_init2 = np.array([th2, omega0])
theta_init3 = np.array([th3, omega0])
theta_init4 = np.array([th4, omega0])

#pass the function deriv to the solver with initial values and integration variable
theta = odeint(deriv,theta_init,t)
theta1 = odeint(deriv,theta_init1, t)
theta2 = odeint(deriv,theta_init2, t)
theta3 = odeint(deriv,theta_init4, t)
theta4 = odeint(deriv,theta_init4, t)

#layering 5 plots of theta vs. time for each initial angle
plt.plot(t,theta[:,0],'g-',label="Numerical solution5")
plt.plot(t,theta1[:,0],'r--',label="Numerical solution4")
plt.plot(t,theta2[:,0],'b-',label="Numerical solution3")
plt.plot(t,theta3[:,0],'r^',label="Numerical solution2")
plt.plot(t,theta4[:,0],'g^',label="Numerical solution1")
plt.xlabel(r'$t (seconds)$',fontsize=20)
plt.ylabel(r'$\theta (rad)$',fontsize=20)
plt.ylim(-1,1.5)
plt.grid()
#creating a space outside the plot for the legend
ax = plt.subplot(111)
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```
