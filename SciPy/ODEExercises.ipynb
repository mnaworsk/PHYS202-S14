## Question 2

(a) Use Euler's method (not odeint) to solve the following first-order differential equation with given initial condition y(0) = 0.
dydx=x+2y
Find the value of y when x = 1, y(1), using two different step sizes: 0.25 then 0.02.

```python
def Euler_method(xmax,dx):
    a = -9.81
    nsteps = int(xmax/dx)
    x = np.zeros(nsteps)
    dydx = np.zeros(nsteps)
    y = np.zeros(nsteps)
    x[0] = 0.
    y[0] = 0.
    for i in range(len(x)-1):
        x[i+1] = x[i] + dx
        dydx[i+1] = x[i] + 2*y[i]
        y[i+1] = y[i]+ dydx[i] * dx
    return x,y

#solve equation using Euler's method
x1,y1 = Euler_method(100,0.25)
x2,y2 = Euler_method(100,0.02)
ii = (x1 == 1)
jj = (x1 == 1)
print y1[ii]
print y2[jj]
```
