### Simple plots

```python
#create NumPy array of independent var (x), 
# then set dependent var (y = f(x)) by using built it NumPy functions

x = np.linspace(0, 2*np.pi, 300)
y = np.sin(x)
y2 = np.sin(x**2)
```

```python
# x is now an array with 300 values ranging from 0 to 2pi. 
#y is the sine array and y2 is the sqaure of the sine array at each of those 300 x values

#now plot it: 
plt.plot(x, y)
plt.plot(x, y2)
plt.show()
```

```python
#controlling style, color, and other propoerties of markers
plt.plot(x, y, linewidth=2);
plt.plot(x, y2, linewidth=2);
```

```python
# decrease the number of points to illustrate the use of markers
x = np.linspace(0, 2*np.pi, 50)
y = np.sin(x)
y2 = np.cos(x)
plt.plot(x, y, 'o', markersize=5, color='r');
plt.plot(x, y2, '^', markersize=5, color='b');
```

### Saving a figure

```python
#back to our original data
x = np.linspace(0, 2*np.pi, 300)
y = np.sin(x)
y2 = np.sin(x**2)

plt.plot(x, y)
plt.plot(x, y2)

#add a grid
plt.grid()

plt.savefig("Example.pdf")
plt.savefig("Example.png")
```

### Refining our plots

```python
#create the data
x = np.linspace(-np.pi, np.pi, 56, endpoint = True)
c, s = np.cos(x), np.sin(x)
```

```python
plt.plot(x, c)
plt.plot(x, s)
plt.show()
```

```python
#Create a figure of size 8X6 points, 80 dots per inch
plt.figure(figsize=(8, 6), dpi=80)

#create a new subplot from a grid of 1x1
plt.subplot(1, 1, 1)

#plot cosine with a blue continuousline of width 1(pixels)
plt.plot(x, c, color="blue", linewidth=1.0, linestyle="-")

#plot sine with a green continuous line of width 1 (pixels)
plt.plot(x, s, color="green", linewidth=1.0, linestyle="-")

#set x limits
plt.xlim(-4.0, 4.0)

#set x ticks
plt.xticks(np.linspace(-1, 1, 5, endpoint=True))

#set y limits
plt.ylim(-1.0, 1.0)

#set y ticks
plt.yticks(np.linspace(-1, 1, 5, endpoint=True))

#save figure using 72 dots per inch
plt.savefig("plot_example.png", dpi=72)

#show results on screen
plt.show()
```

### Setting tick labels

```python
plt.figure(figsize=(10,6), dpi=80)
plt.plot(x, c, color="blue", linewidth=2.5, linestyle="-")
plt.plot(x, s, color="red", linewidth=2.5, linestyle="-")

plt.xlim(x.min() * 1.1, x.max() * 1.1)
plt.ylim(c.min() * 1.1, c.max() * 1.1)

plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi$'])

plt.yticks([-1, 0, +1], [r'$-1$', r'$0$', r'$+1$'])

plt.show()
```

### Moving spines

```python
plt.figure(figsize=(10,6), dpi=80)
plt.plot(x, c, color="blue", linewidth=2.5, linestyle="-")
plt.plot(x, s, color="red", linewidth=2.5, linestyle="-")

plt.xlim(x.min() * 1.1, x.max() * 1.1)
plt.ylim(c.min() * 1.1, c.max() * 1.1)

#moves the spine
ax = plt.gca() #gca- get current axis
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data', 0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data', 0))

plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi$'])

plt.yticks([-1, 0, +1], [r'$-1$', r'$0$', r'$+1$'])

plt.show()
```

### Adding a legend

```python
plt.figure(figsize=(10,6), dpi=80)
plt.plot(x, c, color="blue", linewidth=2.5, linestyle="-")
plt.plot(x, s, color="red", linewidth=2.5, linestyle="-")

plt.xlim(x.min() * 1.1, x.max() * 1.1)
plt.ylim(c.min() * 1.1, c.max() * 1.1)

ax = plt.gca() #gca- get current axis
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data', 0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data', 0))

plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi$'])

plt.yticks([-1, 0, +1], [r'$-1$', r'$0$', r'$+1$'])

#adds a legend in the upper left corner
#the label argument can be added to the plot commands above, rather than retyping them as we did here:
plt.plot(x, c, color="blue", linewidth=2.5, linestyle="-", label="cosine")
plt.plot(x, s, color="red", linewidth=2.5, linestyle="-", label="sine")
plt.legend(loc='upper left')

plt.show()
```

### Annotate some points

```python
plt.figure(figsize=(10,6), dpi=80)
plt.plot(x, c, color="blue", linewidth=2.5, linestyle="-")
plt.plot(x, s, color="red", linewidth=2.5, linestyle="-")

plt.xlim(x.min() * 1.1, x.max() * 1.1)
plt.ylim(c.min() * 1.1, c.max() * 1.1)

ax = plt.gca() #gca- get current axis
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data', 0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data', 0))

plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi$'])

plt.yticks([-1, 0, +1], [r'$-1$', r'$0$', r'$+1$'])

#adds a legend in the upper left corner
#the label argument can be added to the plot commands above, rather than retyping them as we did here:
plt.plot(x, c, color="blue", linewidth=2.5, linestyle="-", label="cosine")
plt.plot(x, s, color="red", linewidth=2.5, linestyle="-", label="sine")
plt.legend(loc='upper left')


#annotate some points
t = 2 * np.pi / 3
plt.plot([t, t], [0, np.cos(t)], color='blue', linewidth=2.5, linestyle="--")
plt.scatter([t, ], [np.cos(t), ], 50, color='blue')

plt.annotate(r'$sin(\frac{2\pi}{3})=\frac{\sqrt{3}}{2}$', xy=(t, np.sin(t)), xycoords='data', xytext=(+10, +30), textcoords='offset points', fontsize=16, arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))

plt.plot([t, t], [0, np.sin(t)], color='red', linewidth=2.5, linestyle="--")
plt.scatter([t, ], [np.sin(t), ], 50, color='red')

plt.annotate(r'$cos(\frac{2\pi}{3})=-\frac{1}{2}$', xy=(t, np.cos(t)), xycoords='data', xytext=(-90, -50), textcoords='offset points', fontsize=16, arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))

plt.show()
```

## Other types of plots

### error bars

```python
x = np.array([0.0, 2.0, 4.0, 6.0, 8.0])
y = np.array([1.1, 1.9, 3.2, 4.0, 5.9])
plt.figure()
plt.errorbar(x, y, xerr=0.2, yerr=0.6, marker='o')
plt.title("Simplest errorbars, 0.2 in x, 0.6 in y");
```

```python
x = np.arange(0.1, 4, 0.5)
y = np.exp(-x)

#variable error bar values
yerr= 0.1 + 0.2*np.sqrt(x)
xerr = 0.1 + yerr

plt.figure()
plt.errorbar(x, y, xerr, yerr, marker='^')
plt.show()
```

### Subplots and logarithmic axes

```python
x = np.linspace(0., 5.)
y = np.exp(-x)

#make figure with 4 subplots and axes side-by-side
fig, ax = plt.subplots(1, 4, figsize=(10, 6))

#plots on each axis
ax[0].plot(x,y)
ax[1].loglog(x,y)
ax[2].semilogx(x,y)
ax[3].semilogy(x,y);
```

### Scatter plots

```python
x = np.arange(0, 100)
y = np.random.rand(100) #100 random numbers

plt.scatter(x,y);
```

### Histograms

```python
#histogram annotated with text inside plot, using text function

mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

#the histogram of the data
n, bins, patches = plt.hist(x, 50, normed=1, facecolor='g', alpha=0.80) #alpha chages the transparency of the fill color

plt.xlabel('Smarts', fontsize=20)
plt.ylabel('Probability', fontsize=20)
plt.title('Histogram of IQ', fontsize=20)

#This will oit a text fragment at the position given
plt.text(45, .027, r'$\mu=100,\ \sigma=15$', fontsize=20)
plt.axis([40, 160, 0, 0.03])
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.grid();
```
